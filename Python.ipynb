```python
#CONTENT ADDED BY MONTH
df2["date_added"] = pd.to_datetime(df2['date_added'])
df2['month_added']=df2['date_added'].dt.month
df2['month_name_added']=df2['date_added'].dt.month_name()
month_order = ['January',
 'February',
 'March',
 'April',
 'May',
 'June',
 'July',
 'August',
 'September',
 'October',
 'November',
 'December']

df2['month_name_added'] = pd.Categorical(df2['month_name_added'], categories=month_order, ordered=True)

data_sub = df2.groupby('type')['month_name_added'].value_counts().unstack().fillna(0).loc[['TV Show','Movie']].cumsum(axis=0).T

fig, ax = plt.subplots(1, 1, figsize=(12, 6))
color = ["#b20710", "#221f1f"]

for i, mtv in enumerate(df2['type'].value_counts().index):
    mtv_rel = data_sub[mtv]
    ax.bar(mtv_rel.index, mtv_rel, color=color[i], label=mtv, alpha=0.9)

ax.yaxis.tick_right()
ax.axhline(y=0, color='black', linewidth=1.3, alpha=0.4)

for s in ['top', 'right', 'bottom', 'left']:
    ax.spines[s].set_visible(False)

ax.set_xticklabels(data_sub.index, fontfamily='serif', rotation=0)
ax.margins(x=0)

fig.text(0.13, 0.95, 'Content added by month [Cumulative Total]', fontsize=15, fontweight='bold', fontfamily='serif')

fig.text(0.13, 0.855, "Movie", fontweight="bold", fontfamily='serif', fontsize=14, color='#b20710')
fig.text(0.19, 0.855, "|", fontweight="bold", fontfamily='serif', fontsize=14, color='black')
fig.text(0.2, 0.855, "TV Show", fontweight="bold", fontfamily='serif', fontsize=14, color='#221f1f')

ax.tick_params(axis=u'both', which=u'both', length=0)

plt.show()
```

```python
#RATING AGES AND GENRES
ratings_ages = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}
df2['target_ages'] = df2['rating'].replace(ratings_ages)
df2['target_ages'].unique()


# Assuming you have a DataFrame named 'edited'
# Split and explode the 'listed_in' column
df = df2.assign(genre=df2['listed_in'].str.split(', ')).explode('genre')

# Calculate the count of each genre
genre_counts = df['genre'].value_counts().reset_index()
genre_counts.columns = ['genre', 'count']
# Get the top 7 genres by count
top_genres = genre_counts.head(7)

# Filter the original DataFrame for genres in the top 7
filtered_df = df[df['genre'].isin(top_genres['genre'])]

# Group by 'genre', 'rating', and count the number of occurrences
result = filtered_df.groupby(['genre', 'target_ages']).size().reset_index(name='count')


# Sort the result by 'count' in descending order
result = result.sort_values(by='count', ascending=False)



pivot_result = result.pivot(index='genre', columns='target_ages', values='count')


pivot_result



```

```python
#COUNTRIES AND GENRES

# Assuming you have a DataFrame named 'edited' with 'country' and 'listed_in' columns

# Extract the top 10 countries based on the count of 'show_id'
top_10_countries = (
    df2['country'].str.split(', ')
    .explode()
    .str.strip()
    .value_counts()
    .head(7)
    .reset_index()
)
top_10_countries.columns = ['Country', 'CountryCount']

# Split and explode the 'listed_in' column
genre_data = (
    df2['listed_in'].str.split(', ')
    .explode()
    .str.strip()
)

# Create a DataFrame to store genre counts for the top 10 countries
genre_counts = pd.DataFrame(columns=['Genre', 'Country', 'GenreCount'])

# Iterate through the top 10 countries
for country in top_10_countries['Country']:
    # Filter the data for the current country
    filtered_data = df2[df2['country'].str.contains(country)]

    # Calculate genre counts for the current country
    country_genre_counts = genre_data[filtered_data.index].value_counts().reset_index().head(5)
    country_genre_counts.columns = ['Genre', 'GenreCount']

    # Add the 'Country' column
    country_genre_counts['Country'] = country

    genre_counts = genre_counts.append(country_genre_counts, ignore_index=True)

# Pivot the result to create a genre-country count matrix
pivot_result = genre_counts.pivot(index='Country', columns='Genre', values='GenreCount')
pivot_result.fillna(0, inplace=True)  # Fill NaN values with 0

# Filter the DataFrame to include only the specified genres
specified_genres = ['Dramas', 'International Movies', 'British TV Shows', 'International TV Shows','Comedies']
filtered_pivot_result = pivot_result[specified_genres]
filtered_pivot_result
```

```python
#DISTRIBUTION OF MOVIE DURATION

# Extract 'duration' and 'type' columns for movies
movie_data = df2[df2['type'] == 'Movie'][['duration', 'type']]

# Extract and round the time duration
movie_data['time_duration'] = movie_data['duration'].str.split(' ').str[0].astype(float).round(2)

# Group by 'time_duration' and count the occurrences
result = movie_data['time_duration'].value_counts().reset_index()
result.columns = ['time_duration', 'count']

# Sort the result by 'time_duration'
result = result.sort_values(by='time_duration')
bar_color = "#b20710"
# Plot the data
plt.figure(figsize=(12, 6))
plt.bar(result['time_duration'],
        result['count'],
        color=bar_color
        )
plt.xlabel('Time Duration',
           fontweight="bold",
          fontfamily='serif',
          fontsize=14,
          color='#b20710')
plt.ylabel('Number of films',
          fontweight="bold",
          fontfamily='serif',
          fontsize=14,
          color='#b20710')
plt.title('Distribution of Movie Time Durations',
          fontweight="bold",
          fontfamily='serif',
          fontsize=14,
          color='#b20710')

x_ticks = [x for x in range(0, 360, 20)]  # Adjust step as needed
plt.xticks(x_ticks,fontname='serif')

# Set custom y-tick values and step
y_ticks = [y for y in range(0, max(result['count']) + 1, 20)]  # Adjust step as needed
plt.yticks(y_ticks,fontname='serif')

plt.show()
```
